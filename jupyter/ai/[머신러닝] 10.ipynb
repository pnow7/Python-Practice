### 03 선형회귀 파이썬 구현 #1 (p.290)

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'D2Coding'
plt.rcParams['axes.unicode_minus'] = False
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0.0, 1.0, 2.0])
y = np.array([3.0, 3.5, 5.5])

w = 0                                                                           # 기울기
b = 0                                                                           # 절편

## Hyper Parameter
lrate = 0.003                                                                   # 학습률
epochs = 500                                                                    # 반복 횟수

n = float(len(X))                                                               # 입력 데이터의 개수
Loss_history = []

# 경사 하강법
for i in range(epochs): 
    y_pred = w*X + b                                                            # 선형 회귀 예측값

    ## Loss (Mean Squared Error: MSE)
    error = y_pred - y                                                          # 오차
    Loss = (error**2).mean()                                                    # Mean Squared Error
    Loss_history.append(Loss)

    print("y_pred:", y_pred, "y: ", y)                                          # 예측값
    print("error: ", error)
    print("Loss: ", Loss)

    ## 현재의 w와 b에서의 Loss를 w와 b로 미분한 값
    dw = (2/n) * sum(X * (y_pred-y))                                            # 넘파이 배열간의 산술 계산은 요소별로 적용
    db = (2/n) * sum(y_pred-y)                                                  # sum()은 모든 요소들의 합을 계산하는 내장 함수

    ## w, b 업데이트(수정) -> Loss가 줄어드는 방향으로
    w = w - lrate * dw                                                          # 기울기 수정
    b = b - lrate * db                                                          # 절편 수정
    
    if i % 100 == 0:
        print (f"epoch={i}  Loss={Loss}    w={w}    b={b}")                     # 기울기와 절편을 출력한다.

        y_pred = w*X + b                                                        # 예측값을 만든다.

        plt.scatter(X, y)                                                       # 점찍기

        plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')     # 예측값은 선그래프로 그린다. 
        plt.show()                                                              
```

```python
Loss_history
```

```python
plt.plot(Loss_history)
plt.title("학습에 따른 Loss(MSE)의 변화")
plt.xlabel('epoch')
plt.ylabel('Loss(MSE)')
plt.show()
```

```python
w, b
```

```python
## 새로운 데이터로 예측하기
## X가 1.5일때 예측값은?
new_X = 1.5
y_pred = w * new_X + b
print("w: ", w, "b: ", b)
print("y_pred: ", y_pred)
```

```python
# 점찍기
plt.scatter(X, y, color = 'pink', label = "원본 데이터")

# 예측값은 선그래프로 그린다. 
# plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')
y_pred = w * X + b
plt.plot(X, y_pred, label = "예측 선")

# 새로운 데이터로 예측한 point
plt.scatter([new_X], [w * new_X + b], color = 'red', label = "새로운 데이터 예측")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model

# 선형회귀 모델 생성
reg = linear_model.LinearRegression()

# 2차원으로 만들어야함(y = x + 3)
X = np.array([[0.0], [1.0], [2.0]])
y = np.array([3.0, 3.5, 5.5])

# 학습을 시킴
reg.fit(X, y)
```

```python
# w값 (계수: coefficient)
reg.coef_
```

```python
# b값 (절편, bias)
reg.intercept_
```

```python
# 모델 성능 평가 (R square) 결정계수: 모델이 얼마나 데이터의 추세를 잘 설명하나
# 1에 가까울 수록 선형성이 높은 데이터(좋은 데이터)
reg.score(X, y)
```

```python
# 새로운 데이터로 예측하기
new_X = [[3.5], [4.5]]
reg.predict(new_X)
```

```python
# 학습데이터와 y값을 산포도로 그림
plt.scatter(X, y, color = 'pink', label = "산포도")

# 학습데이터를 입력으로 예측값 계산
y_pred = reg.predict(X)

# 학습 데이터와 예측값을 선그래프로 
plt.plot(X, y_pred, color = 'blue')
```

### LAB 선형회귀 실습 - p.293

```python
"""
인간의 키와 몸무게는 어느 정도 비례할 것으로 예상된다. 
아래와 같은 데이터가 있을 때, 선형 회귀를 이용하여 학습시키고 
키가 165cm 일 때의 예측값을 얻어보자
"""

import matplotlib.pyplot as plt
from sklearn import linear_model

# 선형 회귀 모델 생성
reg = linear_model.LinearRegression()

## 같은방법
# from sklearn.linear_model import LinearRegression()
# reg = LinearRegression

# X: 키(반드시 2차원으로), y: 몸무게
X = [[174], [152], [138], [128], [186]]
y = [71, 55, 46, 38, 88]

# X 데이터와 y 데이터를 학습시킴(reg.fit)
reg.fit(X, y)

print(reg.predict([[165]]))

# 학습 데이터와 y값을 산포도
plt.scatter(X, y, color = 'black')

# 학습 데이터를 입력으로 하여 예측값을 계산
y_pred = reg.predict(X)

# 학습 데이터와 예측값으로 선그래프로 그린다.
# 계산된 기울기와 y 절편을 가지는 직선이 그려짐
plt.plot(X, y_pred, color = 'blue', linewidth = 3, label = "예측치 직선")
plt.title("키와 몸무게 예측치 직선과 y 절편 예측값")
plt.xlabel("키")
plt.ylabel("몸무게")
plt.legend()
plt.show()
```
